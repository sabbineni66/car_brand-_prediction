```python
from tensorflow.keras.layers import Input,Lambda,Dense,Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
#re_size all the images to this
IMAGE_SIZE=[224,224]

train_path=r"C:\Users\hp\Desktop\gopi\Deeplearning_Project\Datasets\train"
valid_path=r"C:\Users\hp\Desktop\gopi\Deeplearning_Project\Datasets\test"
```

```python
[224,224]+[3]###RGB channel
```

```python
#import the resnet50 library as shown below and add preprocessing  layer to the front of ResNet50
#here we will be using imagenet weights
resnet=ResNet50(input_shape=IMAGE_SIZE +[3],weights='imagenet',include_top=False)
```

```python
resnet.summary()
```

```python
#don't train on existing weights
for layer in resnet.layers:
    layer.trainable=False
```

```python
#useful for getting number of output classes
folders=glob('Datasets/Train/*')

```

```python
folders
```

```python
#our layers -you can add more if you want
x=Flatten()(resnet.output)
```

```python
len(folders)
```

```python
prediction=Dense(len(folders),activation='softmax')(x)

#create model object
model=Model(inputs=resnet.input,outputs=prediction)
```

```python
#view the structure of the model

model.summary()
```

```python
#tell the model what cost and optimization to use
model.compile(
loss='categorical_crossentropy',
optimizer='adam',
metrics=['accuracy']
)
```

```python
#use the image Data Generator to import the images in the dataset

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen=ImageDataGenerator(rescale=1./255,
                                shear_range=0.2,
                                zoom_range=0.2,
                                horizontal_flip=True)

test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
#make sure you provide  some target size as intialized for image size
training_Set=train_datagen.flow_from_directory('Datasets/train',
                                              target_size=(224,224),
                                              batch_size=32,
                                              class_mode='categorical')
```

```python
test_Set=test_datagen.flow_from_directory('Datasets/test',
                                              target_size=(224,224),
                                              batch_size=32,
                                              class_mode='categorical')
```

```python
#fit the model
r=model.fit_generator(training_Set,
                      validation_data=test_Set,
                      epochs=50,
                      steps_per_epoch=len(training_Set),
                      validation_steps=len(test_Set)
                      )
```

```python
r.history
```

```python
#plot the loss
plt.plot(r.history['loss'],label='train_loss')
plt.plot(r.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

#plot the accuracy
plt.plot(r.history['accuracy'],label='train acc')
plt.plot(r.history['val_accuracy'],label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
